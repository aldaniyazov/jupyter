#### Задача 1

1.1. Скачать данные по ссылке https://gbcdn.mrgcdn.ru/uploads/asset/5348082/attachment/bc87260b5f1af8c111f94688b18a69d4.csv
Считать данные с помощью pandas
Вывести на экран первые 5 строк

````python
import pandas as pd
url = 'https://gbcdn.mrgcdn.ru/uploads/asset/5348082/attachment/bc87260b5f1af8c111f94688b18a69d4.csv'
df = pd.read_csv(url)
df.head(5)
````

1.2. Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

````python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df['delta_renovated'] = df['delta_renovated'].fillna(0)
df.head(5)
````

1.3 Создайте признаки года продажи, месяца продажи

````python
# Преобразование столбца "date" в формат даты
df['date'] = pd.to_datetime(df['date'])

# Создание нового признака "year_sold" с годом продажи
df['year_sold'] = df['date'].dt.year

# Создание нового признака "month_sold" с месяцем продажи
df['month_sold'] = df['date'].dt.month

df.head(5)
````

1.4 Удалите признаки date, zipcode, lat, long

````python
df = df.drop(['date', 'zipcode', 'lat', 'long'], axis=1)
df.head(5)
````

#### Задача 3

Составьте несколько сводных таблиц

3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

````python
pivot_table_3_1 = clients.pivot_table(index='bedrooms', values='price', aggfunc='mean').sort_values('price', ascending=True)
print(pivot_table_3_1)
````

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

````python
pivot_table_3_2 = clients.pivot_table(index='condition', values='price', aggfunc=['min', 'mean', 'max'])
print(pivot_table_3_2)
````

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

````python
pivot_table_3_3 = clients.pivot_table(index='waterfront', columns='view', values='house_id', aggfunc='count', fill_value=0)
print(pivot_table_3_3)
````

3.4 Каких домов в зависимости от этажности и количества спален больше?

````python
pivot_table_3_4 = clients.pivot_table(index='floors', columns='bedrooms', values='house_id', aggfunc='count', fill_value=0)
print(pivot_table_3_4)
````

3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

````python
pivot_table_3_5 = clients.pivot_table(index='condition', columns='grade', values='price', aggfunc='median')
print(pivot_table_3_5)
````
